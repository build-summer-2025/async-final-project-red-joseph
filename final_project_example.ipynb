```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
advanced_df = pd.read_csv('Advanced.csv')

# Display the first few rows of the dataframe to ensure it is loaded correctly
# Display different information about the first dataset
display(advanced_df.head())
display(advanced_df.info())
display(advanced_df.describe())
display(advanced_df.isna().sum())
```

```python
# cleaning of first dataset
# keep only relevant columns we used in the questions
cols_to_keep = [
    "season", "player", "player_id", "age", "g", "mp",
    "per", "ts_percent", "ows", "dws", "ws", "ws_48",
    "bpm", "vorp"
]

advanced_clean = advanced_df[cols_to_keep].copy()

# drop rows where all major stats are missing
advanced_clean = advanced_clean.dropna(subset=["per", "ts_percent", "ws", "ws_48", "bpm"])

# reset index
advanced_clean = advanced_clean.reset_index(drop=True)

# show missing values again after cleaning
print(advanced_clean.isna().sum())
display(advanced_clean.head())

```

```python

# Load the second dataset
# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
pt_df = pd.read_csv('Player Totals.csv')

# Display the first few rows of the dataframe to ensure it is loaded correctly
# Display different information about the second dataset
display(pt_df.head())
display(pt_df.info())
display(pt_df.describe())
display(pt_df.isna().sum())
```

```python
# Basic cleaning of pt_df
# keep only relevant columns
cols_to_keep = [
    "season", "player", "player_id", "age", "g", "mp",
    "orb", "drb", "trb", "ast", "stl", "blk", "tov", "pf", "pts"
]

pt_cleaned = pt_df[cols_to_keep].copy()

# fill missing personal fouls with 0
pt_cleaned["pf"] = pt_cleaned["pf"].fillna(0)

# fill count-based stats with 0 where missing (orb, drb too this time!)
count_cols = ["mp", "orb", "drb", "stl", "blk", "tov"]
pt_cleaned[count_cols] = pt_cleaned[count_cols].fillna(0)

# recompute total rebounds as orb + drb (ignores old trb completely)
pt_cleaned["trb"] = pt_cleaned["orb"] + pt_cleaned["drb"]

# drop rows where age is missing (can't be imputed logically)
pt_cleaned = pt_cleaned.dropna(subset=["age"])

# reset index
pt_cleaned = pt_cleaned.reset_index(drop=True)

# show remaining missing values
print(pt_cleaned.isna().sum())
display(pt_cleaned.head())
```

```python
# group by player and sum total career points
career_points = pt_cleaned.groupby("player")["pts"].sum().reset_index()

# sort descending and take top 10 players
top10_points = career_points.sort_values(by="pts", ascending=False).head(10)

#display data frame
display(top10_points)

# plot
plt.figure(figsize=(10,6))
plt.barh(top10_points["player"], top10_points["pts"], color="skyblue")
plt.xlabel("Total Career Points")
plt.ylabel("Player")
plt.title("Top 10 NBA Players by Career Points")
plt.gca().invert_yaxis()  # Highest scorer at the top
plt.show()
```

```python
# group by player and sum total career rebounds
career_rebounds = pt_cleaned.groupby("player")["trb"].sum().reset_index()

# sort descending and take top 5 players
top5_rebounds = career_rebounds.sort_values(by="trb", ascending=False).head(5)

# display the final dataframe
print(top5_rebounds)

# Plot
plt.figure(figsize=(8,5))
plt.barh(top5_rebounds["player"], top5_rebounds["trb"], color="orange")
plt.xlabel("Total Career Rebounds")
plt.ylabel("Player")
plt.title("Top 5 NBA Players by Career Rebounds")
plt.gca().invert_yaxis()  # Highest rebounder at the top
plt.show()
```

```python
# group by player and sum total career assists
career_assists = pt_cleaned.groupby("player")["ast"].sum().reset_index()

# sort descending and take top 10 players
top10_assists = career_assists.sort_values(by="ast", ascending=False).head(10)

# display the final dataframe
print(top10_assists)

# Plot
plt.figure(figsize=(10,6))
plt.barh(top10_assists["player"], top10_assists["ast"], color="lightgreen")
plt.xlabel("Total Career Assists")
plt.ylabel("Player")
plt.title("Top 10 NBA Players by Career Assists")
plt.gca().invert_yaxis()  # Highest at the top
plt.show()
```

```python
# group by player and sum total career Win Shares
career_ws = advanced_clean.groupby("player")["ws"].sum().reset_index()

# sort descending and take top 5 players
top5_ws = career_ws.sort_values(by="ws", ascending=False).head(5)

# display the final dataframe
print(top5_ws)

# plot using seaborn
plt.figure(figsize=(8,5))
sns.barplot(
    data=top5_ws,
    x="ws",
    y="player",
    hue="player",
    dodge=False,
    legend=False,
    palette="viridis"
)
plt.xlabel("Total Career Win Shares (WS)")
plt.ylabel("Player")
plt.title("Top 5 NBA Players by Career Win Shares (WS)")
plt.show()
```

```python
# group by player and calculate average PER
career_per = advanced_clean.groupby("player")["per"].mean().reset_index()

# sort descending and take top 5
top5_per = career_per.sort_values(by="per", ascending=False).head(5)

# display dataframe
print(top5_per)

# plot with seaborn
plt.figure(figsize=(8,5))
sns.barplot(
    data=top5_per,
    x="per",
    y="player",
    hue="player",
    dodge=False,
    legend=False,
    palette="mako"
)
plt.xlabel("Average Career Player Efficiency Rating (PER)")
plt.ylabel("Player")
plt.title("Top 5 NBA Players by Career Average PER")
plt.show()
```

```python
# group by player and calculate average TS%
career_ts = advanced_clean.groupby("player")["ts_percent"].mean().reset_index()

# sort descending and take top 10 players
top10_ts = career_ts.sort_values(by="ts_percent", ascending=False).head(10)

# display dataframe
print(top10_ts)

# plot with seaborn
plt.figure(figsize=(8,6))
sns.barplot(
    data=top10_ts,
    x="ts_percent",
    y="player",
    hue="player",
    dodge=False,
    legend=False,
    palette="crest"
)
plt.xlabel("Average Career True Shooting Percentage (TS%)")
plt.ylabel("Player")
plt.title("Top 10 NBA Players by Career TS%")
plt.show()
```

```python
# group by player and calculate average BPM
career_bpm = advanced_clean.groupby("player")["bpm"].mean().reset_index()

# sort descending and take top 10 players
top10_bpm = career_bpm.sort_values(by="bpm", ascending=False).head(10)

# display dataframe
print(top10_bpm)

# plot with seaborn
plt.figure(figsize=(8,6))
sns.barplot(
    data=top10_bpm,
    x="bpm",
    y="player",
    hue="player",
    dodge=False,
    legend=False,
    palette="rocket"
)
plt.xlabel("average career box plus/minus (BPM)")
plt.ylabel("player")
plt.title("top 10 nba players by career BPM")
plt.show()
```

```python
# group by player to get career games and seasons
career_games = pt_cleaned.groupby("player").agg(
    career_games=("g", "sum"),
    seasons_played=("season", "nunique")
).reset_index()

# sort descending by career games and take top 5
top5_games = career_games.sort_values(by="career_games", ascending=False).head(5)

# display dataframe
print(top5_games)

# plot with seaborn
plt.figure(figsize=(8,5))
sns.barplot(
    data=top5_games,
    x="career_games",
    y="player",
    hue="player",
    dodge=False,
    legend=False,
    palette="flare"
)
plt.xlabel("Career games played")
plt.ylabel("Player")
plt.title("Top 5 nba players by career games played")
plt.show()
```

```python
# group by player and calculate average WS/48
career_ws48 = advanced_clean.groupby("player")["ws_48"].mean().reset_index()

# sort descending and take top 10 players
top10_ws48 = career_ws48.sort_values(by="ws_48", ascending=False).head(10)

# display dataframe
print(top10_ws48)

# plot with seaborn
plt.figure(figsize=(8,6))
sns.barplot(
    data=top10_ws48,
    x="ws_48",
    y="player",
    hue="player",
    dodge=False,
    legend=False,
    palette="coolwarm"
)
plt.xlabel("Win Shares Per 48 Minutes (WS/48)")
plt.ylabel("Player")
plt.title("Top 10 NBA Players by Career WS/48")
plt.show()
```

```python
# get career points from pt_cleaned
career_points = pt_cleaned.groupby("player")["pts"].sum().reset_index()
career_points.rename(columns={"pts": "career_points"}, inplace=True)

# get career win shares from advanced_clean
career_ws = advanced_clean.groupby("player")["ws"].sum().reset_index()
career_ws.rename(columns={"ws": "career_win_shares"}, inplace=True)

# merge both on player
merged = pd.merge(career_points, career_ws, on="player")

# compute correlation
correlation = merged["career_points"].corr(merged["career_win_shares"])
print(f"Correlation between career points and career win shares: {correlation:.3f}")

# plot scatter with regression line
plt.figure(figsize=(8,6))
sns.regplot(
    data=merged,
    x="career_points",
    y="career_win_shares",
    scatter_kws={"alpha":0.6},
    line_kws={"color":"red"}
)
plt.xlabel("Career Points")
plt.ylabel("Career Win Shares")
plt.title("Correlation Between Career Points and Career Win Shares")
plt.show()
```
